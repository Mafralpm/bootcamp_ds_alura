# Aula 3 - Manipulação de Dados

```
import matplotlib.pyplot as plt
dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/agendamento-hospitalar/main/dados/A160324189_28_143_208.csv', sep=';', encoding='ISO-8859-1',
                    skiprows=3, skipfooter=12, thousands='.', decimal=',')

colunas_validas = dados.mean().index.tolist()
colunas_validas.insert(0, 'Unidade da Federação')
dados = dados[colunas_validas]

dados = dados.set_index('Unidade da Federação')
dados.head()
```

```
colunas = dados.T.head().columns.to_list()
```

```
regiao = []
for index in colunas:
  lista_split = index.split()
  numero = lista_split[0][:1]
  if(numero == '1'):
    regiao.append('NO')
  elif(numero == '2'):
    regiao.append('NE')
  elif(numero == '3'):
    regiao.append('SU')
  elif(numero == '4'):
    regiao.append('SD')  
  else:
    regiao.append('CO')

print(regiao)

dados['Região'] = regiao
```

```
dados.T
```

```
sec_dados = dados.loc[dados['Região'] == 'CO']

amapa = dados.T['16 Amapá'].to_list()
print(len(amapa))

sec_dados.loc['16 Amapá'] = amapa

sec_dados = sec_dados.drop(['Total', 'Região'], axis=1)

sec_dados.T.head()
```

```
ax = sec_dados.T.plot(figsize=(12,7))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.legend(ncol=3)
plt.show()
```
